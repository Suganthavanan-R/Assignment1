```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/House Price India.csv')

# Display the summary statistics
print(data.describe())

# Display the column-wise data types
print(data.dtypes)

# Display the number of unique values in each column
print(data.nunique())
```

```python
# Drop rows with missing values
data.dropna(inplace=True)

# Drop columns with missing values
data.dropna(axis=1, inplace=True)
```

```python
# Replace missing values with the mean of the column
data.fillna(data.mean(), inplace=True)

# Replace missing values with the median of the column
data.fillna(data.median(), inplace=True)

# Replace missing values with the mode of the column
data.fillna(data.mode(), inplace=True)

# Replace missing values with a custom value
data.fillna('Unknown', inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/House Price India.csv',nrows=10)

# Create a scatter plot matrix to visualize the relationship between pairs of variables
pd.plotting.scatter_matrix(data, figsize=(12, 12), diagonal='hist')

# Create a heatmap to visualize the correlation between variables
corr_matrix = data.corr()
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.show()
```
